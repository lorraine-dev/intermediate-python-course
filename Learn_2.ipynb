````python
import math 

def is_prime(n):
    if n == 1:
        return False
    elif n <= 3:
        return True
    sqrt_n = math.sqrt(n)
    list_check = [n for n in range(int(sqrt_n)+1)]
    for i in list_check[2:]:
        if n % i == 0:
            return False
    return True
````

````python
def select_student(students, threshold):
  result_dict = { 'Accepted' : [], 'Refused' : [], }
  for student in students:
    if int(student[1]) >= threshold:
      result_dict['Accepted'].append(student)
    else:
      result_dict['Refused'].append(student)
  result_dict['Accepted'].sort(key = lambda x: x[1], reverse = True)
  result_dict['Refused'].sort(key = lambda x: x[1], reverse = False)
  print(result_dict)
````

Find the next friday the 13th.

Write a function named friday_the_13th, which takes no parameter, and just returns the date of the next friday the 13th.

If today is a friday the 13th, return it, not the next one.

Return the date as a string of the following format: YYYY-MM-DD.

````python
from datetime import datetime, timedelta
import calendar
def friday_the_13th():
  datetime_object = datetime.now()
  while not calendar.day_name[datetime_object.weekday()] == "Friday":
    datetime_object = datetime_object + timedelta(days = 1)
  while not datetime_object.day == 13:
    datetime_object = datetime_object + timedelta(days = 7)
  return str(datetime_object.date())

friday_the_13th()
````

Write a simple function, named is_anagram, taking two strings, and returning a boolean value.

The function shall return True if the letters of one word are a rearrangement of the letters of the other, False otherwise.

Superfluous (or missing) spaces, quotes, dashes, ... are allowed: funeral is an anagram of real fun.

Capitalized letter are equivalent to lower case letters: Madam Curie is an anagram of Radium came.

Diactirics of any language are ignored: crâné is an anagram of crane.

````python
import unicodedata

def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    only_ascii = nfkd_form.encode('ASCII', 'ignore')
    return only_ascii

def is_anagram(left, right):
  left_set = set(n for n in list(remove_accents(left.lower())) if n <= 122 and n >= 97) 
  right_set = set(n for n in list(remove_accents(right.lower())) if n <= 122 and n >= 97)
  return left_set == right_set

is_anagram("funeral", "real fun")
````

Write a flatten function, which given a list of lists (of any depth) returns a flattened version of it.

````python
def flatten(a_list):
  final_list = []
  for elem in a_list:
    if isinstance(elem, list):
      final_list.extend(flatten(elem))
    else:
      final_list.append(elem)
  return final_list

  
````

Simulate a perfect suffle of a deck of cards.

A perfect shuffle of a deck of card is splitting a deck of cards into equal halves, and perfectly interleaving them.

Perfectly shuffling [1, 2, 3, 4, 5, 6] gives [1, 4, 2, 5, 3, 6].

(We consider that a deck of cards has an even number of cards.)

You'll expose a perfect_shuffle(deck) function, perfectly shuffling the given iterable.

Did you know that if you shuffle 10 times a deck of 1024 cards, the deck returns to its initial state ? It's probably a good way to test your implementation.

````python
def perfect_shuffle(deck):
  left_half = deck[:int(len(deck)/2)]
  right_half = deck[int(len(deck)/2):]
  interleave = []
  for index in range(len(left_half)):
    interleave.append(left_half[index])
    interleave.append(right_half[index])
  return interleave
````

````python
new_list = list(range(1024))
for iter in range(10):
  new_list = perfect_shuffle(new_list)
  print(new_list)
````

Provide a function that prints the n first numbers of the fibonacci sequence.

Name your function fibonacci and take a single parameter, say n.

Your function should return a list of n values from the sequence.

````python
def fibonacci(count):
  if count == 0: return []
  fib_list = [1]
  while len(fib_list) < count:
    if len(fib_list) >= 2 : elem_1 = fib_list[len(fib_list) - 2] 
    else: elem_1 = 0 
    fib_list.append(elem_1 + fib_list[len(fib_list) - 1])
  return fib_list
````

Create a program to find and print all numbers between from 0 to 1000 (both included), that are divisible by 7 and whose digits sum are divisible by 3.

For instance, 1512 is divisible by 7 and the sum of all is digits is 9 and is divisible by 3.

````python
def sum_of_digits(num):
  start = 10
  sum = 0
  while num > 0:
    sum += num % start
    num = int(num/start)
  return sum
````

````python
input = list(range(1500, 1700))
for num in input:
  if num % 7 == 0 and sum_of_digits(num) % 3 == 0:
    print(num)
````
